# Training model MLP-Concrete Dropoutfor MNIST

```python
from train_model import run_experiment
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.pyplot import figure
from matplotlib.ticker import MaxNLocator
```

```python
converged_dropouts_50000 , training_curve_50000, validation_curve_50000, test_curve_50000 =  \
   run_experiment(train_size = 50000, validation_size = 10000, epochs = 20)
```

```python
#Store results for plotting purposes 
training_curve_50000 = np.asarray(training_curve_50000)
validation_curve_50000 = np.asarray(validation_curve_50000)
test_curve_50000 = np.asarray(test_curve_50000)
np.save('results/training_curve_50000.npy', training_curve_50000 )
np.save('results/validation_curve_50000.npy', validation_curve_50000 )
np.save('results/test_50000.npy', test_curve_50000 )
np.save('results/dropouts_50000.npy', converged_dropouts_50000 )
```

```python
#Plot training loss

ax = figure(figsize = (20,8)).gca()
plt.title("Training Loss vs Epochs (data fraction = 1.0)", fontsize = 20)
plt.plot(training_curve_50000, label = "Train")
plt.plot(validation_curve_50000, label = "Validation")
plt.plot(test_curve_50000, label = "Test")
plt.xlabel('Epochs',fontsize = 15)
plt.ylabel('Loss', fontsize = 15)
ax.xaxis.set_major_locator(MaxNLocator(integer=True))
plt.ylim(0.010, 0.020)  
plt.legend(fontsize = 18)
plt.show()
```

```python
ax = figure().gca()
plt.title("Converged Dropout Rates per Layer (data fraction = 1.0)", fontsize = 15)
plt.xlabel('Layers')
plt.ylabel('Dropout Probability')
plt.scatter(np.arange(4)+1,converged_dropouts_50000, c="red")
ax.xaxis.set_major_locator(MaxNLocator(integer=True))
plt.show()
```

## Evolution of Dropout Rate per Layer

```python
converged_dropouts_50000 , training_curve_50000, validation_curve_50000, test_curve_50000 =  \
   run_experiment(train_size = 50000, validation_size = 10000, epochs = 100)
```

```python
#Evolution of dropout 
d = np.asarray(converged_dropouts_50000)

ax = figure(figsize = (20,8)).gca()
plt.title("Dropout Rate vs Epochs", fontsize = 20)
plt.plot(d[:,0], label = "Layer 1")
plt.plot(d[:,1], label = "Layer 2")
plt.plot(d[:,2], label = "Layer 3")
plt.plot(d[:,3], label = "Layer 4")
plt.xlabel('Epochs',fontsize = 15)
plt.ylabel('Dropout Probability', fontsize = 15)
ax.xaxis.set_major_locator(MaxNLocator(integer=True))
plt.legend(fontsize = 18)
plt.show()
```
