# Training model MLP-Concrete Dropoutfor MNIST

```python
from train_model import run_experiment
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.pyplot import figure
from matplotlib.ticker import MaxNLocator
```

```python
converged_dropouts , training_curve, validation_curve, test_curve =  \
   run_experiment(train_size = 50000, validation_size = 10000, epochs = 20)
```

```python
#Plot training loss
ax = figure().gca()
plt.title("Training Loss vs Epochs", fontsize = 15)
plt.plot(training_curve, label = "Train")
plt.plot(validation_curve, label = "Validation")
ax.xaxis.set_major_locator(MaxNLocator(integer=True))
plt.ylabel("Loss")
plt.xlabel("Epochs")
plt.legend()
plt.show()
```

```python
ax = figure().gca()
plt.title("Converged Dropout Rates per Layer", fontsize = 15)
plt.xlabel('Layers')
plt.ylabel('Dropout Probability')
plt.scatter(np.arange(4)+1,converged_dropouts, c="red")
ax.xaxis.set_major_locator(MaxNLocator(integer=True))
plt.show()
```
